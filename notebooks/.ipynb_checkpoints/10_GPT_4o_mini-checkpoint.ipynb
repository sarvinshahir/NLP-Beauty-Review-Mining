````python
import openai
from openai import OpenAI

client = OpenAI(api_key="YPUR_API_KEY")
````

````python
def get_aspect_tags(review_text):

    prompt = f"""
    Extract ONLY short 1–2 word beauty aspect tags from this review.
    No sentences. No generic words like “foundation” or “cream”.
    Only meaningful product qualities such as:
    hydrating, lightweight, cakey, oxidizes, long-wearing, greasy, drying, fragrance-free.

    Output this JSON format exactly:
    {{
      "benefits": [...],
      "disadvantages": [...]
    }}

    Max 8 tags per section.

    Text:
    {review_text}
    """

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )

    # IMPORTANT: new SDK uses .content, not ["content"]
    return response.choices[0].message.content
````

````python
text = "This foundation is very hydrating but it oxidizes and becomes cakey on oily skin."
print(get_aspect_tags(text))
````

````python
# red color function
def red_color_func(*args, **kwargs):
    return "rgb(200, 0, 0)"   # rich red

wc_negative = WordCloud(
    width=800,
    height=400,
    background_color="white",
    color_func=red_color_func,
    collocations=False
).generate(negative_text)

plt.figure(figsize=(10, 5))
plt.imshow(wc_negative, interpolation="bilinear")
plt.axis("off")
plt.title("Negative Word Cloud (Disadvantages)", fontsize=18)
plt.show()
````

````python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# green color function
def green_color_func(*args, **kwargs):
    return "rgb(0, 150, 0)"   # dark green

wc_positive = WordCloud(
    width=800,
    height=400,
    background_color="white",
    color_func=green_color_func,
    collocations=False
).generate(positive_text)

plt.figure(figsize=(10, 5))
plt.imshow(wc_positive, interpolation="bilinear")
plt.axis("off")
plt.title("Positive Word Cloud (Benefits)", fontsize=18)
plt.show()
````

````python
df_found = final_df[final_df["product_name"].isin(foundations)]
df_moist = final_df[final_df["product_name"].isin(moisturizers)]
````

````python
avg = all_reviews.groupby("product_name")["rating"].mean().reset_index()
avg.columns = ["product_name", "avg_rating"]

avg
````

````python
df_found = avg[avg["product_name"].isin(foundations)]
df_moist = avg[avg["product_name"].isin(moisturizers)]
````

````python
import matplotlib.pyplot as plt
import matplotlib as mpl

# --- Soft, aesthetic pink theme ---
foundation_color = "#F7A8B8"    # baby-pink
moist_color      = "#F4C2C2"    # pastel rose
bg_color         = "#FFF9FB"    # ultra-light pink
text_color       = "#5A4A4A"    # warm neutral brown

plt.rcParams.update({
    "figure.facecolor": bg_color,
    "axes.facecolor": bg_color,
    "savefig.facecolor": bg_color,
    "font.family": "DejaVu Sans",
    "axes.titlesize": 16,
    "axes.labelsize": 14,
    "xtick.labelsize": 12,
    "ytick.labelsize": 12
})

# --- Create figure ---
plt.figure(figsize=(20, 8))

# ========================
#       FOUNDATIONS
# ========================
plt.subplot(1, 2, 1)
plt.barh(
    df_found["product_name"],
    df_found["avg_rating"],
    color=foundation_color,
    edgecolor="#E38DA8",
    linewidth=1.5
)
plt.title("Average Rating — Foundations", color=text_color)
plt.xlabel("Rating (0–5)", color=text_color)
plt.xlim(0, 5)
plt.gca().invert_yaxis()
plt.grid(axis="x", linestyle="--", alpha=0.3, color="#EBC6D3")

# ========================
#      MOISTURIZERS
# ========================
plt.subplot(1, 2, 2)
plt.barh(
    df_moist["product_name"],
    df_moist["avg_rating"],
    color=moist_color,
    edgecolor="#D99AA0",
    linewidth=1.5
)
plt.title("Average Rating — Moisturizers", color=text_color)
plt.xlabel("Rating (0–5)", color=text_color)
plt.xlim(0, 5)
plt.gca().invert_yaxis()
plt.grid(axis="x", linestyle="--", alpha=0.3, color="#EBC6D3")

# ========================
#     Main Poster Title
# ========================
plt.suptitle(
    "Average Product Ratings (Foundations vs Moisturizers)",
    fontsize=22,
    color=text_color,
    y=1.02
)

plt.tight_layout()
plt.show()
````
