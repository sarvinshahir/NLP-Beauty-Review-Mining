```python
import pandas as pd

df = pd.read_csv("all_reviews.csv")
print(df.shape)
df.head()
df.info()
```

```python
# Drop rows with missing or empty reviews
df = df.dropna(subset=['review_text'])
df = df[df['review_text'].str.strip() != '']

print(df.shape)
```

```python
# Average rating and number of reviews per product
product_stats = (
    df.groupby("product_name")
      .agg(avg_rating=('rating', 'mean'),
           num_reviews=('rating', 'count'))
      .sort_values(by='num_reviews', ascending=False)
)
product_stats.head(10)

retailer_stats = (
    df.groupby(['retailer', 'product_name'])
      .agg(avg_rating=('rating', 'mean'),
           num_reviews=('rating', 'count'))
      .reset_index()
)
retailer_stats.head()
```

# Preprocess & Clean Text

```python
import re, string

def clean_text(text):
    text = str(text).lower()
    text = re.sub(f"[{string.punctuation}]", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df["clean_review"] = df["review_text"].apply(clean_text)
```

# Derive Sentiment Labels from Ratings

```python
def rating_to_sentiment(r):
    if r >= 4:
        return "positive"
    elif r == 3:
        return "neutral"
    else:
        return "negative"

df["sentiment"] = df["rating"].apply(rating_to_sentiment)
df["sentiment"].value_counts(normalize=True)
```

```python
df['rating'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
sns.countplot(data=df, x="retailer", hue="sentiment")
plt.title("Sentiment Distribution by Retailer")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(
    df["clean_review"], df["sentiment"], test_size=0.2, random_state=42
)

vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1,2))
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

model = LogisticRegression(max_iter=1000, class_weight='balanced')
model.fit(X_train_tfidf, y_train)
y_pred = model.predict(X_test_tfidf)

print(classification_report(y_test, y_pred))
```
