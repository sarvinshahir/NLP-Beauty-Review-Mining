```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


df = pd.read_csv("products_benefit_disadv.csv")
```

```python
category_map = {
    "Armani Luminous Foundation": "foundation",
    "BYOMA Moisturizing Rich Cream": "moisturizer",
    "Clinique Dramatically Different Moisturizing Lotion+": "moisturizer",
    "Dior Forever Skin Glow": "foundation",
    "Drunk Elephant Lala Retro Whipped Moisturizer": "moisturizer",
    "Estée Lauder Double Wear": "foundation",
    "Estée Lauder Revitalizing Supreme+ Global Anti-Aging Cream": "moisturizer",
    "Fenty Pro Filt'r Soft Matte Foundation": "foundation",
    "Kiehl’s Ultra Facial Moisturizing Cream with Squalane": "moisturizer",
    "Lancôme Rénergie H.P.N. 300-Peptide Cream": "moisturizer",
    "Lancôme Teint Idole Ultra Wear Foundation": "foundation",
    "NARS Light Reflecting Foundation": "foundation"
}

df["category"] = df["product_name"].map(category_map)

```

```python
tfidf = TfidfVectorizer()
X = tfidf.fit_transform(df["benefits"].fillna(""))


sim_matrix = cosine_similarity(X)

sim_df = pd.DataFrame(
    sim_matrix,
    index=df["product_name"],
    columns=df["product_name"]
)

```

```python
df.to_csv("products_final.csv", index=False)
sim_df.to_csv("similarity_matrix.csv")

print("Files created:")
print(" - products_final.csv")
print(" - similarity_matrix.csv")
```
